```python
import pandas as pd
```

```python
sequences = [
    [
        (0, {'a'}),
        (86400, {'a', 'b', 'c'}),
        (259200, {'a', 'c'}),
    ],
    [
        (0, {'a', 'd'}),
        (259200, {'c'}),
    ],
    [
        (0, {'a', 'e', 'f'}),
        (172800, {'a', 'b'}),
    ],
]
sequences
```

```python
sequences = pd.DataFrame(
    {(i, j): (interval, elements)
     for i, sequence in enumerate(sequences)
     for j, (interval, elements) in enumerate(sequence)}, ).transpose()
sequences.columns = ['interval', 'element']
sequences
```

```python
sequences = sequences.explode('element')
sequences
```

```python
sequences.index.names = ['sequence', 'item']
sequences
```

```python
sequences.index.get_level_values('item')
```

```python
interval = 0
element = 'a'
itemize = lambda t: t // 86400
```

```python
sequences['itemized_interval'] = sequences['interval'].apply(itemize)
sequences
```

```python
sequences[sequences['element'] == element]['itemized_interval'].unique()
```

```python
sequences[(sequences['interval']==interval)&(sequences['element']==element)]
```

```python
itemize = lambda i: i // 86400
sequences['itemized_value'] = sequences['interval'].apply(itemize)
sequences
```

```python
supports = sequences.groupby(level=0)['element'].apply(set).explode().value_counts()
supports
```

```python
counts = sequences['element'].value_counts()
counts
```
